#### Refined Function

```python
def color_of_da():
    """Assigning a color to the grade of a D&A student within the 5.0 range
    """
    #Declaring/initialiizing a list
    avg_of_da = []
    #input for number of students
    n = int(input("How many students? "))
    #loop for n input
    for i in range(n):
        #Declare the score input
        avg = float(input("What is your average score: "))
        #Appending the score to the initialized list
        avg_of_da.append(avg)
    #loop to assign colors to elements in the list     
    for s in avg_of_da:
        #Condition to assign/print colors based on score 
        if s <= 3.0 and s > 0:
            print("Red")
        elif s <= 5.0 and s > 3.0:
            print("Green")
        else:
            print("Grade is Invalid")

color_of_da()
```

### Variable-length arguments
##### *args
* The *args syntax allows a function to accept a non-finite number of non-keyword arguments.
* When *args is used in a function definition, all the extra arguments passed to the function are collected into a tuple.

##### *kwargs
* The **kwargs syntax allows a function to accept a non-finite number number of keyword arguments (arguments passed as key=value).
* When **kwargs is used in a function definition, all the extra keyword arguments passed to the function are collected into a dictionary.

```python
def show_args(*args, **kwargs):
    print("args:", args)
    print("kwargs:", kwargs)

show_args(1, 2, 3, 5, 6, 7, "name", name="John", age=25, ball="Big G")
```

```python
#Function to check if a number is odd or even
def check_even():
    n = int(input("Enter number: "))
    if (n % 2 == 0):
        print(f"{n} is even")
    else:
        print(f"{n} is odd")

check_even()
```

### Control Flow with Functions
We often combine functions with loops and conditionals.

```python
# if, elif, else
def check_number(n):
    if n > 0:
        print("Positive")
    elif n < 0:
        print("Negative")
    else:
        print("Zero", "\n")

check_number(10)
check_number(-5)
check_number(0)

# For loop with function
def print_list(lst):
    for item in lst:
        print(item)
    print("\n")

print_list([1, 2, 3, 4])

# While loop with function
def countdown(n):
    while n > 0:
        print(n)
        n -= 1
    print("Done!", "\n")

countdown(5)

# break, continue, pass
def loop_demo():
    for i in range(5):
        if i == 2:
            continue  # Skip 2
        if i == 4:
            break     # Stop loop
        if i == 1:
            pass      # Do nothing
        print(i)

loop_demo()
```
