## DATA STRUCTURES

### LISTS
* The items in a list are enclosed in a square bracket
* The items can also be of different datatypes
* A list is ordered
* A list is mutable (you can change the contents)
* Each data element in a list should be separated by a comma

```python
my_list = [1, 3, 'Annie', 5.0]
my_list
```

```python
my_list2 = [9, 4, 5.3, 1]
my_list2
```

```python
my_list2[2]
```

```python
my_list[0:3]
```

```python
my_list2[2] = 5.3
my_list2
```

```python
my_list2[-1] = 298 #-1 is the last value in your list
my_list2
```

```python
my_list2[1:4]
```

```python
my_list2[1:4] = 6, 4, 87
my_list2
```

### SETS
* The items in a set are enclosed in a curly bracket
* The items can also be of different datatypes
* A set is unordered
* A set is immutable (you cannot change the contents) and cannot be indexed
* Each data element in a set should be separated by a comma
* A set cannot be duplicated

```python
my_set = {1, 9, 5, 'Blessing', 3, 5, 23, 56}
my_set
#presence of a string turns everything to a string
#looking at the first "number"
```

```python
my_set[2]
```

```python
len(my_set)
```

```python
my_set2 = {1, 9, 5, 3, 5, 23, 56}
my_set2
#sorts like normal numbers without the presence of a string
```

```python
my_set3 = {1, 3, 2, 5, 8, 4, True}
#Boolean True = 1 and False = 0
my_set3
```

```python
my_set4 = {3, 2, 5, 8, 4, True}
my_set4
```
