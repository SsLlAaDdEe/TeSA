## Control FLow
#### Conditional Statements 
* if
* elif
* else

if condition:
    execute code
elif condition:
    execute code
else:
    execute code

```python
if True:
    print("True")
elif False:
    print("False")
```

```python
if True:
    print("True")
else:
    print("False")
```

```python
if 5 < 10:
    print("5 is less than 10")
elif 5 == 10:
    print("5 is equal than 10")
else:
    print("5 is greater than 10")
```

```python
if 5 in range(1, 10):
    print("5 is less than 10")    
```
