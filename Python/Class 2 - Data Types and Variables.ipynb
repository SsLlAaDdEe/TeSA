## VARIABLES
* It is best practice for our variable names to be in small letters
* Variable names cannot start with numbers or special characters
* Variable names can only start with an alphabet or an underscore (_)
* Variable names cannot contain special characters, aside from underscores
* Variable names cannot contain white space; they can be joined with an underscore
* Variable names can be alphanumeric, but with a letter starting them
* It's best practice to give your variables descriptive names
* While declaring variable names, it is not to be put in quotes (" ") like a string
* Do not use built-in functions or Python reserved keywords (true or false) in variable names

```python
name = "Favour"
```
