#### To call a variable in the print function
* Comma
* Concatenation
* F-string
* str.format()

```python
name = "Arogundade"
```

```python
print(name, "is a student of TeSA cohort 4")
#Comma
```

```python
print(name + " is a student of TeSA cohort 4")
#Concatenation
```

```python
print(f"{name} is a student of TeSA cohort 4, and {name} is a smart student")
#F-string
```

```python
print("{} is a student of TeSA cohort 4, and {} is a smart student".format(name, name))
#str.format()
```

```python
name1 = input("What is your name? ")
score = input("Enter your TeSA technical score: ")
print("The technical score for ", name1, " is ", score)
#anything returned is a string
```

```python
name1 = str(input("What is your name? "))
score = float(input("Enter your TeSA technical score: "))
print("The technical score for ", name1, " is ", score)
#int will be converted to float
```
