### Indexing

Major Rules
* The first character is at index 0, the second at 1, and so on.
* The last character of a string is at index n-1, where n is the length of the string
* Indexing can also start from the last character, with an index of -1

```python
a = "hilda"
```

```python
a[0]
```

```python
a[4]
```

```python
grade = 'average'
```

```python
len(grade)
```

```python
grade[4]
```

### Slicing

Rules
* It is done using the format `[start:end]`
* The slice includes the `start` index, but stops just before the `end` index.

```python
name= "Emmanuel"
```

```python
name[0:6]
```

[position of first value : number of characters + position of first value ] 

[position of first value : 1 + position of last value ]

```python
name[0:4]
```

```python
name[4:8]
```

```python
name[0:8]
```

```python
name[2:5]
```

```python
s=[1,2,3,4]
s[-1:-1]=[5,6,7]
```

```python
s
```

```python
y=[1,2,3,4]
y[:0]=[5,6,7]
```

```python
y
```

## Lists

* The items in a list are enclosed in a square bracket
* The items in a list can be of different datatypes
* A list is ordered
* A list is mutable


```python
my_list = [1,'Emmanuel', 1.0]
my_list
```

```python
my_list2 = [2,3,1,4,2,2]
my_list2
```

```python
my_list2[2]
```

```python
my_list2[0:4]
```

```python
my_list2[2] = 5
my_list2
```

```python
my_list2[-1] = 0
my_list2
```

```python
my_list2[1:4]
```

```python
my_list2[6:9] = 1,2,3
my_list2
```
