## Control Flow

### Looping Statements (for, while)

```python
for i in range (1,71):
    if i % 5 == 0:
        print(f"{i} is divisible by 5")
    elif i % 7 == 0:
        print(f"{i} is divisible by 7")
else:
    print("Have a lovely day")
```

```python
i = 1
while i < 71:
    if i % 5 == 0:
        print(f"{i} is divisible by 5")
    elif i % 7 == 0:
        print(f"{i} is divisible by 7")
    i += 1
```

```python
my_list = ['Seyi', 'Enjeghe', 'Kamara', 'Daniel', 'Tega']
name = 0
# Print "name is a TeSA student"
while name < len(my_list):
    print(f"{my_list[name]} is a TeSA student")
    name += 1
```

```python
for name in my_list:
    print(f"{name} is a TeSA student")
```

```python
for i in my_list[0]:
    for n in i[0]:
        print(n)
```

```python
count = 0
while count < 5:
    count+=1
    if count == 3:
        break
    print(count)
    
```

```python
count = 0
number = 7
while count < 10:
    count+=1
    if count == number:
        continue
        print(f"Number {number} found")
    print(f"current number {count}")
```

```python
num = 5
count = num
while count > 0:
    fact = num * (num - 1)
    count-=1
    num-=num
    print(fact)
```
