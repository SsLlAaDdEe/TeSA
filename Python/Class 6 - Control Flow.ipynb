## Control Flow

### Conditional Statements (if, elif, else)
* if statement will ONLY return a result if condition is true
* elif statement will ONLY return a result if previous condition(s) is/are false, and this is true

```python
if True:
    print('x')
```

```python
if False:
    print('y')
#if statement will ONLY return a result if condition is true
```

```python
if 5 < 2:
    print('Correct! 5 is less than 2')
elif 5 > 2:
    print('Correct! 5 is greater than 2')
```
