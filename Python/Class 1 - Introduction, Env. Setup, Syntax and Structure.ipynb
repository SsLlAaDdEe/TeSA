# Python Shortcuts
## Python Shortcuts
### Python Shortcuts
* Esc + M = Change cell style from Code to Markdown
* Esc + Y = Change cell style from Markdown to Code
* Esc + A = To add a cell right above the cell you're in
* Esc + B = To add a cell right below the cell you're in
* Esc + DD = To delete a cell
* Shift + Enter = To execute a cell
* **To make bold**
* *To make italics*
* ***To make both bold and italics***
